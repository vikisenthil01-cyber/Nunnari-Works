```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/nunnarilabs/ml/master/house/kc_house_data.csv"
df = pd.read_csv(url)

print(df.head())
print(df.info())

print(df.isnull().sum())
df = df.dropna()
print(df.isnull().sum())

corr = df.corr(numeric_only=True)
print(corr)

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

plt.scatter(df["sqft_living"], df["price"])
plt.xlabel("sqft_living")
plt.ylabel("price")
plt.title("sqft_living vs price")
plt.show()

df2 = sns.load_dataset("titanic")

print(df2.isnull().sum())

df2["age"] = df2["age"].fillna(df2["age"].mean())
df2["embarked"] = df2["embarked"].fillna(df2["embarked"].mode()[0])

print(df2.isnull().sum())

df2 = df2.drop(columns=["deck"])
df2 = df2.drop(columns=["alive", "class"])

print(df2.columns)

df2 = df2.rename(columns={
    "pclass": "Passenger_Class",
    "sibsp": "Siblings_Spouses",
    "parch": "Parents_Children"
})

print(df2.columns)

df2["age"] = df2["age"].astype(int)
df2["survived"] = df2["survived"].astype(bool)
df2["Passenger_Class"] = df2["Passenger_Class"].astype("category")

print(df2.dtypes)

url2 = "https://raw.githubusercontent.com/selva86/datasets/master/Auto.csv"
df3 = pd.read_csv(url2)

print(df3.columns)

print(df3["origin"].value_counts())
print(df3["cylinders"].value_counts())
print(df3["year"].value_counts(normalize=True))
print(df3["name"].value_counts().head())

print(df3.groupby("origin")["mpg"].mean().sort_values(ascending=False))
print(df3.groupby("cylinders")["horsepower"].mean())
print(df3.groupby("origin")["mpg"].mean())

print(df3.groupby("cylinders")["mpg"].agg(["min","max","mean"]))

corr3 = df3.corr(numeric_only=True)
print(corr3)

print(corr3["mpg"].sort_values(ascending=False))

print(df3["horsepower"].corr(df3["mpg"]))

plt.scatter(df3["horsepower"], df3["mpg"])
plt.xlabel("horsepower")
plt.ylabel("mpg")
plt.title("Horsepower vs MPG")
plt.show()

sns.scatterplot(x="weight", y="mpg", data=df3)
plt.show()

sns.scatterplot(x="horsepower", y="mpg", hue="origin", data=df3)
plt.show()

plt.hist(df3["mpg"], bins=20)
plt.title("MPG Distribution")
plt.show()

sns.histplot(df3["horsepower"], kde=True)
plt.show()

sns.histplot(df3["weight"])
plt.show()

plt.figure(figsize=(12,8))
sns.heatmap(corr3, annot=True, cmap="coolwarm")
plt.show()
```
