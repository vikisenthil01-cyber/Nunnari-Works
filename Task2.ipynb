```python
movies = ["Inception", "Interstellar", "Avatar", "Hangover", "Joker"]

# first & last
print(movies[0], movies[-1])

# add & remove
movies.append("Matrix")
movies.pop(1)
print(movies)

# sort
movies.sort()
print(movies)

# length & check
print(len(movies))
print("Avatar" in movies)
```

```python
cities = ("Chennai", "Salem", "Mumbai", "Bangalore", "Delhi")

print(cities[1], cities[3])

# cities[0] = "Madurai"  # ❌ Error: tuple is immutable

# unpacking
a, b, c, d, e = cities
print(a, b, c, d, e)

# convert
city_list = list(cities)
city_list.append("Madurai")
cities = tuple(city_list)
print(cities)
```

```python
weekday = {"Math", "Physics", "Python"}
weekend = {"Python", "AI", "ML"}

print(weekday | weekend)   # union
print(weekday & weekend)   # intersection
print(weekday - weekend)   # difference

weekday.add("English")
weekend.remove("ML")

print("AI" in weekday or "AI" in weekend)
```

```python
def greet(name):
    print(f"Hello, {name}! Welcome to Day 2.")

def add_numbers(a, b):
    return a + b

def find_max(a, b, c):
    return max(a, b, c)

def is_even(n):
    return n % 2 == 0

greet("Vignesh")
print(add_numbers(5, 10))
print(find_max(3, 7, 5))
print(is_even(4))
```

```python
def introduce(name, language="Python"):
    print(f"{name} is learning {language}")

introduce("Vignesh")
introduce("Vignesh", "SQL")

def calculate_discount(price, discount=10):
    return price - (price * discount / 100)

print(calculate_discount(1000))
```

```python
num = int(input("Enter number: "))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")

score = int(input("Enter score: "))
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")

year = int(input("Enter year: "))
if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print("Leap Year")
else:
    print("Not Leap Year")
```

```python
# 1 to 20
for i in range(1, 21):
    print(i)

# even numbers
for i in range(2, 51, 2):
    print(i)

# fruits with index
fruits = ["apple", "banana", "cherry"]
for i in range(len(fruits)):
    print(i, fruits[i])

# sum 1 to 100
total = 0
for i in range(1, 101):
    total += i
print(total)

# vowels count
text = "hello world"
count = 0
for ch in text:
    if ch in "aeiou":
        count += 1
print(count)
```
